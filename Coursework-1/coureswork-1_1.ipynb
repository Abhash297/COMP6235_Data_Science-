```python
import numpy as np
from scipy import stats

sample_data = df['Time'] 
mean = np.mean(sample_data)


confidence_interval = stats.t.interval(0.10, len(sample_data) - 1, loc=np.mean(sample_data), scale=stats.sem(sample_data))

print(f"Sample Mean: {mean}")
print(f"95% Confidence Interval Time: {confidence_interval}")
```
